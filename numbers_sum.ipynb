```python
Problem: 
    There is an array of numbers without duplicates . 
    Write a code to find if there exists 2 numbers that sums upto a number given.
```

```python
Example : [ 1,2,3,4,5,6,7]  Sum= 9 .. There exists 2 and 7
    
    
    There is brute force method to travers every number with every  other number . But this is not optimal since it will take
    O(n^2) time.
```

```python
Solution : Hash-Table. Will use extra memory but will be faster.
    
    Hash tables are a type of data structure in which the address or the index value of the data element is generated from 
    a hash function. That makes accessing the data faster as the index value behaves as a key for the data value. In other 
    words Hash table stores key-value pairs but the key is generated through a hashing function.
    
So the search and insertion function of a data element becomes much faster as the key values themselves become the
index of the array which stores the data.
In Python, the Dictionary data types represent the implementation of hash tables. 
The Keys in the dictionary satisfy the following requirements.
The keys of the dictionary are hashable i.e. the are generated by hashing function which generates unique result for each 
unique value supplied to the hash function.
The order of data elements in a dictionary is not fixed.
```

```python
METHOD 3 :
      USING LEFT POINTER AND RIGHT POINTER :
            
    Step 1 : The array must be sorted in increasing order first.
    Step 2 : Pick a left pointer and right pointer at 2 extreme ends.
    Step 3 : Add the 2 pointers' values and check if it matches the sum.
    Step 4 : If the added value is more than sum and since the array is sorted, if right pointer is moved towards the left
             the added sum will decrease further. Hence the left pointer must be moved by 1 spot. 
            Vice -Versa till the sum is found.
            
            
            Time : o(nlogn)
            space : o(1)
```

```python
#method 1  : o(n^2) time


def numbersum(array,targetsum):
    for i in range(len(array)-1):   #because there is no number the last number so no need to check for the last one.
        firstnum=array[i]
        for j in range(len(array)):
            secondnum=array[j]
            if secondnum+firstnum==targetsum:
                return [firstnum,secondnum]
            
    return
array=[1,2,3,4,5,6,7]
targetsum=16
result=numbersum(array,targetsum)
print(result)
```

```python
#method 2 : o(n) time

def numbersum(array,targetsum):
    nums={}
    for i in range(len(array)):
        num=array[i]
        if targetsum-num in nums:
            return [num, targetsum-num]
        else:
            nums[num]=True   # Instead of True can be anything..        
    return

array=[1,20,13,40,35,61,77]
targetsum=78
result=numbersum(array,targetsum)
print(result)
```

```python
#method 3 : o(nlogn)

def numbersum(array,targetsum):
    array.sort()
    left=0
    right=len(array)-1
    while left<right:
        currentsum=array[left]+array[right]
        if currentsum==targetsum:
            return [array[left],array[right]]
        if currentsum<targetsum:
            left=left+1
        elif currentsum > targetsum:
            right=right-1
    return

array=[1,20,13,40,35,61,77]
targetsum=33
result=numbersum(array,targetsum)
print(result)           
```
